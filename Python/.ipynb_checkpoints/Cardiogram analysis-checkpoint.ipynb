## Чтение и запись данных

```python
data_train = loadtxt('train.csv', delimiter=',', dtype=int, skiprows=1)
data_test = loadtxt('test.csv', delimiter=',', dtype=int, skiprows=1)
X = data_train[:, 1:]    # train features
y = data_train[:, 0]     # train target (class)
Xt = data_test[:, 1:]    # test features
```

Введем функцию, которая будем записывать в файл полученную классификацию в нужном формате.

```python
def write_predictions(filename, yt):
    with open(filename, 'w') as f:
        f.write('id,target\n')
        for id, target in enumerate(yt):
            f.write('%d,%f\n' % (id, target))
```

## Метод ближайшего соседа

```python
from sklearn import neighbors
```

Найдем оптимальные параметры для метода ближайших соседей. Для этого заведем соответствующую функцию:

```python
from sklearn.grid_search import GridSearchCV


def plot_optim_param(clf,params,X,y):
    grid_searcher = GridSearchCV(clf, params,n_jobs=-1,scoring='roc_auc') # не забываем правильно указать критерий оценки
    grid_searcher.fit(X, y)
    
    print grid_searcher.best_estimator_
    
    #график
    means = []
    stds = []
    
    for elem in grid_searcher.grid_scores_:
        means.append(np.mean(elem[-1]))
        stds.append(sqrt(np.var(elem[-1])))
    means = np.array(means)
    stds = np.array(stds)
    
    pl.figure(figsize=(8, 6))
    pl.plot(rng,means)
    pl.fill_between(rng, means + stds, means - stds, alpha = 0.1, facecolor='blue')
    pl.xlabel(str(params.keys()[0]))
    pl.ylabel('AUC') 
```

```python
rng = range(1,101, 5)
clf = neighbors.KNeighborsClassifier(warn_on_equidistant=False)
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

```python
rng = range(10,31)
clf = neighbors.KNeighborsClassifier(warn_on_equidistant=False)
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

Попробуем использовать найденное значение 25 соседей и обучим классификатор 

```python
clfKNN = neighbors.KNeighborsClassifier(warn_on_equidistant=False, n_neighbors=25)
clfKNN.fit(X, y)
yt = clfKNN.predict_proba(Xt)
write_predictions('submission_knn25.csv', yt[:, 1])
```

Результат на тестовой выборке: 0.90161

## Линейный SVM

Теперь воспользуемся линейным SVM:

```python
from sklearn import svm
```

Введем функцию оценивания на части учебной выборки

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
def make_test(clf,X,y):  
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
       
    clf.fit(X_train, y_train)
    y_pred =  clf.predict_proba(X_test)[:, 1]
    
   
    return y_pred, y_test, roc_auc_score(y_test,y_pred)
```

```python
def make_n_tests(clf,X,y,n):  
    auc_scores = []
    for i in range(n):
        y_pred, y_test, auc_score = make_test(clf,X,y)
        auc_scores += [auc_score]
    
    return mean(auc_scores), var(auc_scores)        
```

Прогоним линейный SVM:

```python
clfSVM = svm.SVC(kernel = 'linear', probability = True)
```

```python
y_pred, y_test, auc_score = make_test(clfSVM,X,y)
```

```python
print auc_score
```

Теперь несколько раз:

```python
auc_av, auc_var = make_n_tests(clfSVM,X,y,15)
```

```python
auc_av
```

Это пока не лучше того, что мы получили с помощью метода ближайших соседей. Попробуем поварьировать параметр С.

```python
rng = [2**(i) for i in range(-10,11)]
params = {'C': rng}
plot_optim_param(clfSVM,params,X,y)
```

Такая маленькая вариативность графика может означать, что все самое интересное происходит за границами рассматриваемого промежутка. Поскольку как оптимальное мы получаем минимальное из значений С, сдвинем интервал влево.

```python
rng = [2**(-i) for i in range(10,25)]
params = {'C': rng}
plot_optim_param(clfSVM,params,X,y)
```

```python
rng = [i*0.00005 for i in range(2,11)]
params = {'C': rng}
plot_optim_param(clfSVM,params,X,y)
```

Обучим классификатор с линейным SVM и найденным параметром C = 0.00025.

```python
clfLSVM = svm.SVC(C = 0.00025, kernel = 'linear', probability = True)
clfLSVM.fit(X, y)
yt = clfLSVM.predict_proba(Xt)
write_predictions('submission_lsvm.csv', yt[:, 1])
```

Полученный результат - 0.90186, что незначительно больше предыдущего результата.

## SVM с полиномиальным ядром

Попробуем еще одно популярное ядро - полиномиальное.

```python
clfPoly = svm.SVC(kernel = 'poly', probability = True)
```

Протестируем его:

```python
make_n_tests(clfPoly,X,y,15)
```

Для полиномиального ядра мы можем варьировать целый ряд параметров - степень многочлена, ядерный коэффициент $\gamma$, независимый коэффициент многочлена.

```python
rng = [i*0.000025 for i in range(1,20)]
params = {'gamma': rng}
plot_optim_param(clfPoly,params,X,y)
```

Оптимальное значение - такое же, какое было найдено для RBF: $\gamma$ = 0.0001.

Теперь попробуем изменить степень многочлена:

```python
rng = range(1,8)
params = {'degree': rng}
plot_optim_param(clfPoly,params,X,y)
```

По умолчанию степень многочлена была равна 3, непохоже, чтобы был смысл отходить от нее.

Попробуем то же самое, но с ненулевым свободным членом многочлена:

```python
clfPoly0 = svm.SVC(kernel = 'poly', probability = True, coef0 = 1)
```

```python
rng = [10**(-i) for i in range(3,10)]
params = {'gamma': rng}
plot_optim_param(clfPoly0,params,X,y)
```

```python
rng = range(1,8)
params = {'degree': rng}
plot_optim_param(clfPoly0,params,X,y)
```

Теперь, когда мы знаем предполагаемые значимые промежутки параметров, будем изменять их одновременно:

```python
C_range = 10.0 ** np.arange(-2, 3)
gamma_range = [i*0.00005 for i in range(1,5)]
degree_range = [1,2,3]
coef0_range = [0,1]
params = dict(gamma=gamma_range, C=C_range, degree = degree_range, coef0 = coef0_range)
grid = GridSearchCV(clfPoly, params)
grid.fit(X, y)

print grid.best_estimator_
```

```python
print grid.best_score_
```

К сожалению, полиномиальное ядро не принесло нам результат, который был бы лучше, чем RBF SVM.
