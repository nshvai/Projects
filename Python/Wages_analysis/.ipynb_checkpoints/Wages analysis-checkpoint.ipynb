Как и в лабораторной работе - используем метод ближайших соседей, используя выбранные параметры.

Сформируем новый датафрейм, состоящий из отобраных признаков

```python
from copy import deepcopy

df_newtr = deepcopy(df_train[f_list])
```

Вспомогательная функция для разворачивания категориальных признаков:

```python
def dummy_encode_categorical_columns(df):
    result_df = pd.DataFrame()
    
    for s in list(df.columns.values):
        if type(df[s][0]) == str:
            dummy = pd.get_dummies(df[s], prefix = s, prefix_sep=': ') 
            result_df = result_df.join(dummy,how = 'right')
        else:
            result_df = result_df.join(df[s],how = 'right')        

    return result_df
```

Dummy-кодируем и нормируем признаки, в отдельный массив запишем метки классов, обозначив их как 0 и 1.

```python
X = dummy_encode_categorical_columns(df_newtr)

from sklearn.preprocessing import scale as scale
X = scale(X)

y = pd.Categorical.from_array(df_train.CLASS).labels
```

Найдем оптимальные параметры для метода ближайших соседей. Для этого заведем соответствующую функцию:

```python
from sklearn.grid_search import GridSearchCV


def plot_optim_param(clf,params,X,y):
    grid_searcher = GridSearchCV(clf, params)
    grid_searcher.fit(X, y)
    
    print grid_searcher.best_estimator_
    
    #график
    means = []
    stds = []
    
    for elem in grid_searcher.grid_scores_:
        means.append(np.mean(elem[-1]))
        stds.append(sqrt(np.var(elem[-1])))
    means = np.array(means)
    stds = np.array(stds)
    
    pl.figure(figsize=(8, 6))
    pl.plot(rng,means)
    pl.fill_between(rng, means + stds, means - stds, alpha = 0.1, facecolor='blue')
    pl.xlabel(str(params.keys()[0]))
    pl.ylabel('Accuracy')         
```

```python
rng = range(1, 100,5)
clf = KNN(warn_on_equidistant=False, weights = 'uniform')
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

```python
rng = range(1, 100,5)
clf = KNN(warn_on_equidistant=False, weights = 'distance')
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

Оптимальным будет значение weights = uniform. Найдем теперь оптимальное количество соседей, уменьшив шаг и сузив отрезок рассмотрения.

```python
rng = range(15, 40)
clf = KNN(warn_on_equidistant=False, weights = 'uniform')
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

Оптимальное количество соседей определим пока как 27, и будем от него отталкиваться.
Теперь поэкспериментируем со степенью в метрике Минковского (это достаточно затратно по времени, поэтому количество соседей мы зафиксировали)

```python
rng = range(10,22)
clf = KNN(warn_on_equidistant=False, n_neighbors=27)
params = {'p': rng}
plot_optim_param(clf,params,X,y)
```

Остановимся на p = 17 (хотя судя по графику, разница несущественна). Прогоним еще раз классификатор по количеству соседей

```python
rng = range(1, 102,5)
clf = KNN(warn_on_equidistant=False, p=17)
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

```python
rng = range(20, 31,1)
clf = KNN(warn_on_equidistant=False, p=17)
params = {'n_neighbors': rng}
plot_optim_param(clf,params,X,y)
```

На гисторамме видны зоны "неуверенности" вокруг возле крайних точек.
"Огрубим" наш классификатор - подтянем области на гистограмме, близкие к 0 и 2 (это правильно угаданные метки), к крайним значениям.

```python
def make_rough(y_pred, catch):
    y_rough = []
    for elem in y_pred:
        if elem < catch:
            y_rough+=[0]
        elif elem > 1-catch:
            y_rough+= [1]
        else:
            y_rough+= [elem]
    return y_rough
```

```python
y_pred_rough = make_rough(y_pred,0.25)
y_err = y_pred_rough + y_test
hist(y_err, bins = 20)
error = MAE(y_pred_rough,y_test)
print "%0.2f" % error
```

```python
X_list = dummy_encode_categorical_columns(df_train[best_list])
X_list = scale(X_list)

clfKNN = KNN(warn_on_equidistant=False, p = 2,n_neighbors = 7)

def make_test_pred(X_list,clf):
    
    X_train, X_test, y_train, y_test = train_test_split(X_list, 
                                                    y, test_size=0.3)

    clf.fit(X_train, y_train)
    y_pred =  clf.predict_proba(X_test)[:, 1]
    
    return y_pred,y_test
```

```python
def find_catch(X_list,clf,rng):
    y_pred,y_test = make_test_pred(X_list,clf)      
    min_err = 1
    
    for catch in rng:
        y_pred_rough = make_rough(y_pred,catch)
        error = MAE(y_pred_rough,y_test)
        
        if error < min_err:
            min_err = error
            opt_catch = catch
        
    return min_err, opt_catch
```

```python
from collections import Counter

error = []
catch = []
rng = [0+0.01*x for x in range(40,51)]

for i in range(0,20):
    min_err, opt_catch = find_catch(X_list,clfKNN,rng)
    error+=[min_err]
    catch+=[opt_catch]
    
catch_count = Counter(catch)
    
print mean(error), catch_count.most_common()
```

Похоже, что выбранные нами признаки - не самый эффективный набор. Попробуем выбрать их в цикле таким образом, чтобы на каждом шаге они максимально улучшали работу решающего дерева.

```python
all_list = df_train.columns[:-1].values.tolist()
```

```python
flag = True
best_list = []
min_err = 1 

while flag:
    
    flag = False
        
    for s in all_list:
        try_list = best_list +[s]
        
        X_list = dummy_encode_categorical_columns(df_train[try_list])
        X_train, X_test, y_train, y_test = train_test_split(X_list, 
                                                    y, test_size=0.3)
        clf = DTC()
        clf.fit(X_train, y_train)
        y_pred =  clf.predict_proba(X_test)[:, 1]
        
        error = MAE(y_pred,y_test) 
        
        if error < min_err:
            best_list = best_list + [s]
            all_list.remove(s)
            min_err = error
            flag = True       
        

print best_list, "%0.2f" % min_err
```

Можно попробовать "скрестить" классификаторы, заставляя их исправлять друг друга в "зонах неуверенности". Будем отдавать приоритет решающему дереву, так как этот классификатор показал себя лучше.

```python
def make_rough_double (y_KNN,y_DTC,catch_KNN,catch_DTC):
    y_rough = []
    for i in range(len(y_KNN)):
        if y_DTC[i] < catch_DTC:
            y_rough+=[0]
        elif y_DTC[i] > 1-catch_DTC:
            y_rough+= [1]
        elif y_KNN[i] < catch_KNN:
            y_rough+=[0]
        elif y_KNN[i] > 1-catch_KNN:
            y_rough+= [1] 
        elif y_DTC[i]>0.5 and y_KNN[i]>0.5:
            y_rough+= [max(y_DTC[i],y_KNN[i])]
        elif y_DTC[i]<0.5 and y_KNN[i]<0.5:
            y_rough+= [min(y_DTC[i],y_KNN[i])]
        
        else:
            y_rough+= [mean([y_DTC[i],y_KNN[i]])]
    return y_rough
```

```python
X_list = dummy_encode_categorical_columns(df_train[best_list])
clfKNN = KNN(warn_on_equidistant=False, n_neighbors = 7)
clfDTC = DTC(max_depth=10,min_samples_leaf = 3)

def make_test_double():  
    X_train, X_test, y_train, y_test = train_test_split(X_list, y, test_size=0.3)
    X_train_sc = scale(X_train)
    X_test_sc = scale(X_test)
    
    clfKNN.fit(X_train_sc, y_train)
    y_predKNN =  clfKNN.predict_proba(X_test_sc)[:, 1]
    clfDTC.fit(X_train, y_train)
    y_predDTC =  clfDTC.predict_proba(X_test)[:, 1]
    
    return y_predKNN,y_predDTC, y_test
```

```python
def find_catch_double(rngKNN,rngDTC):
    y_predKNN,y_predDTC, y_test = make_test_double()      
    min_err = 1
    
    for catchKNN in rngKNN:
        for catchDTC in rngDTC:
            y_pred_rough = make_rough_double(y_predKNN,y_predDTC,catchKNN,catchDTC)
            error = MAE(y_pred_rough,y_test)
            if error < min_err:
                min_err = error
                opt_KNN,opt_DTC = catchKNN,catchDTC
        
    return min_err, opt_KNN,opt_DTC
```

Стоит попробовать. Будем подбирать коэффициенты смешивания и огрубления.

```python
def find_mix(rng_a,rng_catch):
    y_predKNN,y_predDTC, y_test = make_test_double()      
    min_err = 1
    
    for a in rng_a:
        for catch in rng_catch:
            y_mix = make_rough(a*y_predKNN + (1-a)*y_predDTC,catch)
            error = MAE(y_mix,y_test)
            if error < min_err:
                min_err = error
                opt_a,opt_catch = a,catch
        
    return min_err, opt_a,opt_catch 
```

```python
error = []
a_list = []
catch_list = []
rng_a = [0.39,0.4,0.41,0.42] # интервалы сузили после нескольких итераций
rng_catch = [0.46,0.47,0.48] # при близких значениях к 0.5 оптимальный весовой коэффициент а начинает стремиться к крайним значениям

for i in range(0,10):
    min_err, opt_a,opt_catch  = find_mix(rng_a,rng_catch)
    error+=[min_err]
    a_list+=[opt_a]
    catch_list+=[opt_catch]

   
a_count = Counter(a_list)
catch_count = Counter(catch_list)
    
print mean(error), mean(a_list), mean(catch_list)
print a_count.most_common()
print catch_count.most_common()
```
